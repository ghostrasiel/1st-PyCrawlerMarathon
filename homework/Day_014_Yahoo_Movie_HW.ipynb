```python
import requests
import bs4
import re

url='https://movies.yahoo.com.tw/movie_intheaters.html'
r = requests.get(url)
reson = r.text
soup = bs4.BeautifulSoup(reson , 'html.parser')

rs = soup.find(class_='page_numbox')
print('共{x}頁'.format(x = len(rs.ul.find_all('a'))))
max_num =len(rs.ul.find_all('a'))
for page_num in range(1 , max_num+1) :
    payload ={'page': str(page_num)}
    a = requests.get(url , params=payload)
    ason = a.text
    aoup = bs4.BeautifulSoup(ason , 'html.parser')
    ps = aoup.find_all('div' , class_='release_info')
    print('第{x}頁'.format(x= page_num))
    for p in ps :
      print(p.find(class_='gabtn').text.strip())
      print('期待值:',p.find(class_='leveltext').span.text)
      print('評價:',p.find(class_='count')['data-num'])
      print(p.find(class_='release_text').text.strip())
      print('----------')
```
