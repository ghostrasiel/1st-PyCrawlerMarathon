```python
import bs4
import requests
import re
def Wikiput(input_keyword , revise):
    if revise <= max_revise :
        utf8_url =repr(input_keyword.encode('UTF-8')).upper()
        # print(utf8_url)
        utf8_url=utf8_url.replace('\X' , '%') #替換\X 成 %
        # print(utf8_url)
        # print("%s:%s"%(input_keyword,utf8_url[2:-1:1]))
        new_utf8=utf8_url[2:-1:1]
        url = 'https://zh.wikipedia.org/wiki/{X}'.format(X = new_utf8)
        # print(url)
        hearders = {
            'authority': 'zh.wikipedia.org',
            'method': 'GET',
            'path': '/wiki/' + new_utf8,
            'scheme': 'https',
            'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3',
            'accept-encoding': 'gzip, deflate, br',
            'accept-language': 'zh-TW,zh;q=0.9,en-US;q=0.8,en;q=0.7,zh-CN;q=0.6',
            'cookie': 'GeoIP=TW:TPE:Taipei:25.05:121.53:v4; TBLkisOn=0; mwPhp7Seed=8b8; WMF-Last-Access-Global=04-Jun-2019; WMF-Last-Access=04-Jun-2019',
            'dnt': '1',
            # 'if-modified-since': 'Tue, 04 Jun 2019 12:03:22 GMT',
            'referer': 'https://zh.wikipedia.org/wiki/Wikipedia:%E9%A6%96%E9%A1%B5',
            'upgrade-insecure-requests': '1',
            'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'
        }
        r = requests.get(url , headers =hearders)
        reson = r.text
        soup = bs4.BeautifulSoup(reson , 'html.parser')
        ps = soup.find_all('p')
        cover=[]
        for p in ps :
            cover.append(p.text)

        str1="".join(cover)
        with open('./WikiArticle/{K}'.format(K=input_keyword) , 'w' , encoding='utf-8') as RS :
            RS.write(str1)
        acovers=[]
        for ext_link in ps:
            a_tag =ext_link.find_all('a',href=re.compile("/wiki/.+"))
            for a in a_tag:
                if len(a.text) > 1 :
                    # print(a.text )
                    acovers.append(a.text)
                    # print(len(a.text))

        # print(acovers)

        w = revise + 1
        print('遞[{U}]圈-{Q} 網址:{F}'.format(U = revise , Q =input_keyword,F=url))
        for acover in acovers :
            Wikiput(acover , w)

p = input()
max_revise = 1
Wikiput(p,0)
```
