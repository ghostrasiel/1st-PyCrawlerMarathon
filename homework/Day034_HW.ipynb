```python
from bs4 import BeautifulSoup
import requests
import random
import time
from selenium import webdriver
url = 'https://free-proxy-list.net/'
delar = webdriver.Chrome('C:/Users/ghost/Desktop/chromedriver.exe')
delar.get(url)
time.sleep(2)
html = delar.page_source
soup = BeautifulSoup(html , 'html.parser')

a = []

for i in range(20):
    try:
        tsoups = soup.find('table').find('tbody').find_all('tr', class_='odd')
        for tsoup in tsoups:
            a.append(tsoup.find('td').text)
        asoups=soup.find('table').find('tbody').find_all('tr',class_='even')
        for asoup in asoups:
            a.append(asoup.find('td').text)
        delar.find_element_by_xpath('//*[@id="proxylisttable_next"]/a').click()

    except:
        break
print(a)
delar.close()
```

```python
for i in range(5):
    ip = random.choice(a)
    print('Use', ip)
    try:
        resp = requests.get('http://ip.filefab.com/index.php',
                            proxies={'http': ip,'https':ip},timeout=10)
        soup = BeautifulSoup(resp.text, 'html5lib')
        print(soup.find('h1', id='ipd').text.strip())
    except:
        print("fial")
```
