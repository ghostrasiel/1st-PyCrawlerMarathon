```python
import requests
url = 'https://www.zhihu.com/explore'
r =requests.get(url)
reson = r.text
print(reson)
```

```python
import requests
url = 'https://www.zhihu.com/explore'
header = {
'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'
}
r =requests.get(url , headers =header)
reson = r.text
print(reson)
```
