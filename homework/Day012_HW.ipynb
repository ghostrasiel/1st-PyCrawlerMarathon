```python
import requests
import bs4
import re

url='https://www.ettoday.net/news/news-list.htm'
r =requests.get(url  ,headers={'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.132 Safari/537.36"})
reson =r.text
soup = bs4.BeautifulSoup(reson , 'html.parser')
href_a=re.compile('/news/20191209/.+')
```

```python
for date in soup.find(class_='part_list_2').find_all('h3') :
    print(date.find(class_='date').text,date.find_all('a')[0].text)
```

```python
href_a=re.compile('/news/news-list-2019-12-09-.+.htm')
manus =soup.find_all('a',href=href_a)
a = 0
for a in range(len(manus)):
    for date in soup.find(class_='part_list_2').find_all('h3') :
        if date.find_all('em')[0].text == manus[a].string:
            print(date.find(class_='date').text,date.find_all('em')[0].text,date.find_all('a')[0].text)
```
