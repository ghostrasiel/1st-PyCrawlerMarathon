```python
import requests
import bs4
import datetime
url='https://www.ptt.cc/bbs/NBA/index.html'
r = requests.get(url)
reson = r.text
soup = bs4.BeautifulSoup(reson , 'html.parser')
today = datetime.datetime.now() #拿取現在的時間
# print(today.month)
for tt in soup.find_all(class_='r-ent'):
    if '刪除' not in tt.find(class_='title').text :
        if str(today.month) in tt.find(class_='date').text:
            print(tt.find(class_='author').text,tt.find(class_='title').text,tt.find(class_='date').text)
            print('----------------------')
```
