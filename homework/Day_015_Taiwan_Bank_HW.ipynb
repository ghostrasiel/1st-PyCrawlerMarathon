```python
import requests
import bs4
import re
url = 'https://rate.bot.com.tw/xrt?Lang=zh-TW'
r= requests.get(url,verify = False)
reson = r.text
soup = bs4.BeautifulSoup(reson , 'html.parser')
r_table =soup.find(name='table',attrs={'title':'牌告匯率'}).find(name='tbody').find_all('tr')
# print(r_table[0])
for r_tab in r_table:
    r_usa_num = r_tab.find(name='div',attrs={'class':'visible-phone print_hide'})
    print(r_usa_num.text.strip())
    r_usa_put =r_tab.find(name='td',attrs={'data-table':'本行現金買入'})
    r_usa_out =r_tab.find(name='td',attrs={'data-table':'本行現金賣出'})
    print('現金買入{x}\n現金賣出{y}'.format(x=r_usa_put.text , y = r_usa_out.text))
    print('------------')
```

```python
from _datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
for a_tab in r_table:
    r_x = a_tab.find(name='td',attrs={'data-table':'歷史匯率'})
    # print(r_x.a['href'])
    r_x_link='https://rate.bot.com.tw/'+r_x.a['href']
    # print(r_x_link)
    good_link=r_x_link.replace('history','quote/2019-11')
    # print(good_link)
    a = requests.get(good_link , verify = False)
    apson =a.text
    asoup = bs4.BeautifulSoup(apson , 'html.parser')
    a_table =asoup.find(name='table' , attrs={'class':'table table-striped table-bordered table-condensed table-hover'}).find('tbody').find_all('tr')
    # print(a_table)
    data_money = list()
    put_money=list()
    out_money=list()
    print(a_table[0].find(name='td',attrs={'class':'text-center tablet_hide'}).text)
    for a_tab in a_table:
        data_string=a_tab.find(name='td',attrs={'class':'text-center'}).text
        data = datetime.strptime(data_string , '%Y/%M/%d').strftime('%Y/%M/%d') #轉換日期
        data_money.append(data)

        put_money.append(a_tab.find_all(name='td',attrs={'class':'rate-content-cash text-right print_table-cell'})[0].text)
        out_money.append(a_tab.find_all(name='td',attrs={'class':'rate-content-cash text-right print_table-cell'})[1].text)
    money_df =pd.DataFrame({'date':data_money , 'put':put_money , 'out':out_money})
    money_df =money_df.set_index('date') #指定欄位為index
    money_df =money_df.sort_index(ascending=True)  #排序
    # print(money_df)
    plt.figure(figsize=(10 , 8))
    x = [1,2,3,1]
    y=[5,6,7,8]
    plt.plot(money_df['put']) #x=data y=put,out
    plt.plot(money_df['out'],linestyle='--')
    plt.legend(loc='upper right')
    plt.show()
    print('---------------')
```
